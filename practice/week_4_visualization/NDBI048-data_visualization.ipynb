# Absolute minimum one should put on the plot

```python
fig, ax = plt.subplots()

ax.hist(df['age'], label='data', rwidth=.95, bins=np.arange(0, 80, 5))
ax.set(
    xlabel='age',  # ax.set_xlabel()
    ylabel='count'  # ax.set_ylabel()
)

ax.set_title("The distribution of passenger's age", fontsize=18)

ax.axvline(df['age'].quantile(.25), c='salmon', label='.25 age')
ax.axvline(df['age'].mean(), c='red', label='avg age')
ax.axvline(df['age'].quantile(.75), c='tomato', label='.75 age')
ax.legend()

ax.set_yticks([10, 100])
ax.set_yticklabels(['10**1', '10**2'])

ax.annotate('mean', (30, 110))
pass
# we can export figure to file
# fig.savefig('simple.png')
```

## some fancier functions

```python
# combination of plots
fig, ax = plt.subplots(1 ,2, figsize=(12, 4))

ax[0].plot([0, 1], [0, 1], label='plot1')
ax[1].plot([0, 1], [10, 0], label='plot2')
```

```python
# secondary axis
fig, ax = plt.subplots()
ax2 = ax.twinx()

ax.plot([0, 1], [0, 1], label='plot1')
ax2.plot([0, 1], [10, 0], label='plot2', marker='o')

ax.legend()
```
